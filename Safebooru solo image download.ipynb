```python
import pandas as pd
import os
import csv
from tqdm import tqdm
```

```python
df = pd.read_csv('data/solo_art.csv')
df.head(5)
```

```python
def makedir(path):
    try:
        os.mkdir(path)
    except:
        print("Directory '", path, "' already exist")
```

```python
image_dir = 'image'
boy = 'boys'
girl = 'girls'
boy_dir = os.path.join(image_dir, boy)
girl_dir = os.path.join(image_dir, girl)
```

```python
both = []
nan = []
art_boy = []
art_girl = []
for idx in tqdm(range(len(df)), ncols=100):
    row = df.iloc[idx]
    tags = row["tags"].split(' ')
    if '1girl' in tags and '1boy' in tags:
        both.append(row.tolist())
    elif '1girl' not in tags and '1boy' not in tags:
        nan.append(row.tolist())
    else:
        if '1girl' in tags:
            art_girl.append(row.tolist())
        else:
            art_boy.append(row.tolist())
print("Hemaphrodite count: ", len(both))
print("Questionable count: ", len(nan))
print("Boys count: ", len(art_boy))
print("Girls count: ", len(art_girl))
print("total: ", len(both) + len(nan) + len(art_boy) + len(art_girl))
print(both[0][-1])
print(nan[0][-1])
print(art_boy[0][-1])
print(art_girl[0][-1])
```

```python
header = df.columns.values.tolist()
#print(header)
with open('data/hemaphrodite.csv', 'w', encoding='utf-8') as data:
    writer = csv.writer(data)
    writer.writerow(header)
    for item in tqdm(both, desc="writing hemaphrodite progress", ncols=100):
        writer.writerow(item)

with open('data/non_gender.csv', 'w', encoding='utf-8') as data:
    writer = csv.writer(data)
    writer.writerow(header)
    for item in tqdm(nan, desc="writing non gender progress", ncols=100):
        writer.writerow(item)

with open('data/boy_art.csv', 'w', encoding='utf-8') as data:
    writer = csv.writer(data)
    writer.writerow(header)
    for item in tqdm(art_boy, desc="writing boy progress", ncols=100):
        writer.writerow(item)

with open('data/girl_art.csv', 'w', encoding='utf-8') as data:
    writer = csv.writer(data)
    writer.writerow(header)
    for item in tqdm(art_girl, desc="writing girl progress", ncols=100):
        writer.writerow(item)
```
